```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

```

```python
df= pd.read_csv('6-bank_customers.csv')# Scale edilmi≈ü bir dataset
```

```python
df.head()
```

```python
df.info()
```

```python
X= df.drop("subscribed",axis=1)
y=df["subscribed"]
```

```python
X.head(
)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=15)
```

```python
from sklearn.linear_model import LogisticRegression
logistic=LogisticRegression()
logistic.fit(X_train,y_train)
y_pred=logistic.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
score= accuracy_score(y_pred,y_test)
print("score:",score)
print(classification_report(y_pred,y_test))
print("confusion matrix:\n",confusion_matrix(y_pred,y_test))# bu TP TN FP FN kƒ±sƒ±mlarƒ±
```

**Hyperparameter tuning**==> Logistic Regression da parametreleri default olarak kullandƒ±k ≈üimdi kendimiz belirleyeceƒüiz

[Logistic Regression](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html)

```python
model=LogisticRegression()
```

```python
penalty=["l1","l2","elasticnet"]
c_values=[100,10,1,0.1,0.01] # C penalty yani cezalandƒ±rmalarƒ±n lambda deƒüeri i√ßin verilen deƒüerdir
solver=["newton-cg","lbfgs","liblinear","newton-cholesky","sag","saga"]
```

```python
params=dict(penalty=penalty,C=c_values,solver=solver)# ≈üimdi yukarƒ±daki deƒüerlerden hangisinin bize en uygun olduƒüunu bulmak i√ßin √ßaƒüƒ±racaƒüƒ±mƒ±z GridSearchCV k√ºt√ºphanesi i√ßin dict olmasƒ± lazƒ±m
```

```python
params
```

**grid searc cv**

```python
from sklearn.model_selection import GridSearchCV,StratifiedKFold
```

```python
cv=StratifiedKFold()# cross validation olu≈üturduk default olarak n_split=5 olur
```

```python
grid=GridSearchCV(estimator=model,param_grid=params,cv=cv,scoring="accuracy",n_jobs=-1)
```

[GridSearchCV](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html)   ==> burdan dok√ºmantasyona bakabiliriz scoring neye g√∂re bize hangi deƒüerlerin en uygun olduƒüunu g√∂stermesini istediƒüimiz b√∂l√ºm

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_ # hangi parametreler benim i√ßin en ideal
```

```python
grid.best_score_
```

```python
y_pred=grid.predict(X_test)
```

```python
score= accuracy_score(y_pred,y_test)
print("score:",score)
print(classification_report(y_pred,y_test))
print("confusion matrix:\n",confusion_matrix(y_pred,y_test))
```

**Random search cv**==> k√ºt√ºphane
[RandomizeSearchCV](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.RandomizedSearchCV.html)

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
models=LogisticRegression()
```

```python
randomcv=RandomizedSearchCV(estimator=models,param_distributions=params,cv=5,scoring="accuracy",n_jobs=-1,n_iter=10)
#n_iter => ka√ß deneme yapƒ±lacaƒüƒ±nƒ± belirler
```

```python
randomcv.fit(X_train,y_train)
```

```python
randomcv.best_params_
```

```python
randomcv.best_score_
```

```python
y_pred=randomcv.predict(X_test)
```

```python
score= accuracy_score(y_pred,y_test)
print("score:",score)
print(classification_report(y_pred,y_test))
print("confusion matrix:\n",confusion_matrix(y_pred,y_test))
```
