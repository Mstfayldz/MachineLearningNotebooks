**Redundancy** ==>Redundancy = aynı bilgiyi tekrar eden değişkenler
Veri setinde bazı değişkenler birbirine çok benzer (hatta neredeyse aynı) bilgi taşırsa, bu fazlalık (redundancy) oluşturur.



**Multicollinearity**, birden fazla bağımsız değişkenin (feature/özellik) birbirleriyle yüksek korelasyona sahip olması durumudur.

```python
# %85 ve üstü korelasyon varsa çıkarılabilir %95 ve üsütyse kessin çıkraılmalıdır
```

```python
print(X_train.corr().iloc[0,2])
```

```python
def correlation_for_dropping(df,threshold): # threshold bizim vereceğimiz yüzde yani onun üstünde bir yüzde ile corr gelirse diye
    columns_drop_set=set()
    corr=df.corr()
    for i in range(len(corr.columns)):
        for j in range(i):                            # bu fonksiyon kalıcı olarak kalabilir ve diğer kodlamalrda da kullanılabilir
          if abs(corr.iloc[i,j])>threshold:
            columns_drop_set.add(corr.columns[i])
    return columns_drop_set
```

```python
correlation_drop=correlation_for_dropping(X_train,0.85)
```

```python
correlation_drop
```

```python
X_train.drop(columns=correlation_drop,inplace=True)#X_train.drop(correlation_drop,axis=1, inplace=True)
X_test.drop(columns=correlation_drop,inplace=True)
```

```python
mse=mean_squared_error(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
print("Mean Squared Error:",mse)
print("Mean Absoulte Error:", mae)
```
